```python
import pandas as pd
from urllib.request import Request, urlopen
from bs4 import BeautifulSoup, Comment
from time import sleep
import re
```

```python
array = []

url = 'https://www.mascotdb.com/native-american-high-school?field_state_tid=All'
hdr = {'User-Agent':'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_2) AppleWebKit/601.3.9 (KHTML, like Gecko) Version/9.0.2 Safari/601.3.9'}
req = Request(url,headers=hdr)
page = urlopen(req).read()
soup = BeautifulSoup(page)
table = soup.find('tbody').find_all('tr')

for tr in table:
    team_name = tr.find('a').text.replace("\n","").replace("  ","")
    url = tr.find('a')['href']
    city = tr.find_all('td')[1].text.replace("\n","").replace("  ","")
    state = tr.find_all('td')[2].text.replace("\n","").replace("  ","")
    status = tr.find_all('td')[3].text.replace("\n","").replace("  ","")
    array.append([team_name,city,state,status,url])
```

```python
nu_array = []
for i in array:
    url = 'https://www.mascotdb.com' + i[4]
    req = Request(url,headers=hdr)
    page = urlopen(req).read()
    soup = BeautifulSoup(page)
    try:
        tidbit = soup.find('strong', text = re.compile('Tidbits: ')).next_sibling
    except:
        tidbit = ''
    nu_array.append([i[0],i[1],i[2],i[3],i[4],tidbit])
```

```python
df = pd.DataFrame(data=nu_array,columns=["team_name", "city","state","status","url",'tidbit'])
```

```python
df.to_csv('allMascotData.csv')
```
