```python
import pandas as pd
```

```python
df = pd.read_csv('allMascotData.csv')
```

```python
active = df.loc[df['status'] == 'active']
```

```python
indians_active = active[active['team_name'].str.contains("Indian", na=False)]
redskins_active = active[active['team_name'].str.contains("Redskin", na=False)]
warriors_active = active[active['team_name'].str.contains("Warriors", na=False)]
braves_active = active[active['team_name'].str.contains("Braves", na=False)]
blackhawks_active = active[active['team_name'].str.contains("Blackhawks", na=False)]
chiefs_active = active[active['team_name'].str.contains("Chiefs", na=False)]
```

```python
active_controversy = pd.concat([indians_active, redskins_active, warriors_active, braves_active, blackhawks_active, chiefs_active])
```

```python
active_controversy = active_controversy.drop(active_controversy.columns[0], axis=1)
```

```python
active_controversy.to_csv('allActive_raw.csv')
```
